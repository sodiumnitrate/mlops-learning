```python
import numpy as np

dt = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
diff_in_mins = []
for el in dt:
    diff_in_mins.append(el.total_seconds()/60)
```

```python
df['duration'] = diff_in_mins
```
